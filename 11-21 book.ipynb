```python
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("USVideos.csv")

newdf = df.sort_values(by = 'views', ascending = False)
newdf.drop_duplicates(subset = 'title', keep = 'first', inplace = True)

#df.drop(df[df['Age'] < 25].index, inplace = True) 

n = newdf.drop(newdf[(newdf['comments_disabled'] == True) & ((newdf["likes"]+newdf['dislikes'])>0)].index)

newdf['category_id'].value_counts()

one = newdf['likes'].sum() / (newdf['likes'] + newdf['dislikes']).sum()
two = newdf['comment_count'].sum() / newdf['views'].sum()
three = (newdf['likes'] + newdf['dislikes']).sum() / newdf['views'].sum()

final = [[one, two, three]]

total = pd.DataFrame(final, columns=['Total Likes to Thumb', 'Total Comments to Views', 'Total Thumbs to Views'])
total

n['comments_disabled'].count()
```

```python
n.loc[(df["category_id"] == 1),"category_name"] = 'Film and Animation'
n.loc[(df["category_id"] == 2),"category_name"] = 'Cars and Vehicles'
n.loc[(df["category_id"] == 10),"category_name"] = 'Music'
n.loc[(df["category_id"] == 15),"category_name"] = 'Pets and Animals'
n.loc[(df["category_id"] == 17),"category_name"] = 'Sport'
n.loc[(df["category_id"] == 19),"category_name"] = 'Travel and Events'
n.loc[(df["category_id"] == 20),"category_name"] = 'Gaming'
n.loc[(df["category_id"] == 22),"category_name"] = 'People and Blogs'
n.loc[(df["category_id"] == 23),"category_name"] = 'Comedy'
n.loc[(df["category_id"] == 24),"category_name"] = 'Entertainment'
n.loc[(df["category_id"] == 25),"category_name"] = 'News and Politics'
n.loc[(df["category_id"] == 26),"category_name"] = 'How to and Style'
n.loc[(df["category_id"] == 27),"category_name"] = 'Education'
n.loc[(df["category_id"] == 28),"category_name"] = 'Science and Technology'
n.loc[(df["category_id"] == 29),"category_name"] = 'Non Profits and Activism'
 
#newdf['category_name'].value_counts()
nudf = pd.DataFrame(n['category_name'].value_counts())

indexnames = nudf[nudf['category_name'] < 150].index
nudf.drop(indexnames, inplace = True)

n['category_name'].value_counts()
```

```python
n['Likes to Thumbs'] = newdf['likes'] / (newdf['likes'] + newdf['dislikes'])

n['Comments to Views'] = newdf['comment_count'] / newdf['views']

n['thumbs to views'] = (newdf['likes'] + newdf['dislikes']) / newdf['views']

dff = n.groupby('category_name').mean()

ratiodf = dff[['Likes to Thumbs' , 'Comments to Views', "thumbs to views"]]
ratiodf
#final = pd.concat([nudf, ratiodf], axis = 1, sort = False)
#final

# da = newdf.groupby('category_name')['dislikes'].max()
# da = pd.DataFrame(da)
# da
n['comments_disabled'].value_counts()
```
