```python
USVids["Likes to Thumbs"] = USVids["likes"] / (USVids["likes"] + USVids["dislikes"])
```

```python
USVids.drop_duplicates(subset = 'video_id', keep = 'last', inplace = True)
#USVids.head()
```

```python
USVids["Comments to Views"] = USVids["comment_count"] / USVids["views"]
#USVids
```

```python
USVids["Thumbs to Views"] = (USVids["likes"] + USVids["dislikes"]) / USVids["views"]
USVids["Thumbs to Views"].describe()
```

```python
df = USVids.groupby("category_name").mean()
```
